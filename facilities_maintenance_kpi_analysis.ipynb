# Facilities Maintenance KPI Analysis

Este notebook analiza órdenes de trabajo de mantenimiento para crear indicadores operativos y detectar oportunidades de mejora.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/maintenance_work_orders.csv')
df['created_at'] = pd.to_datetime(df['created_at'])
df['completed_at'] = pd.to_datetime(df['completed_at'])
df.head()
```

```python
df.info()
df.describe()
```

## KPIs principales

```python
total_orders = len(df)
sla_rate = df['sla_met'].mean()
avg_resolution = df['resolution_hours'].mean()
total_cost = df['total_cost'].sum()
print('Total de órdenes:', total_orders)
print('Cumplimiento SLA:', round(sla_rate, 3))
print('Tiempo promedio de resolución:', round(avg_resolution, 1), 'horas')
print('Costo total estimado:', round(total_cost, 2))
```

## Órdenes de trabajo por categoría

```python
category_counts = df.groupby('category')['work_order_id'].count().sort_values()
category_counts
```

```python
plt.figure(figsize=(8,5))
category_counts.plot(kind='barh')
plt.title('Órdenes de trabajo por categoría')
plt.xlabel('Cantidad de órdenes')
plt.ylabel('Categoría')
plt.tight_layout()
plt.show()
```

## Cumplimiento SLA por prioridad

```python
priority_order = ['Low','Medium','High','Critical']
sla_priority = df.groupby('priority')['sla_met'].mean().reindex(priority_order)
sla_priority
```

```python
plt.figure(figsize=(8,5))
sla_priority.plot(kind='bar')
plt.title('Cumplimiento SLA por prioridad')
plt.xlabel('Prioridad')
plt.ylabel('Cumplimiento SLA')
plt.tight_layout()
plt.show()
```

## Costo mensual de mantenimiento

```python
monthly_cost = df.groupby('month')['total_cost'].sum()
monthly_cost
```

```python
plt.figure(figsize=(8,5))
monthly_cost.plot(kind='line', marker='o')
plt.title('Costo mensual de mantenimiento')
plt.xlabel('Mes')
plt.ylabel('Costo total estimado')
plt.tight_layout()
plt.show()
```

## Horas de paro por área

```python
downtime_site = df.groupby('site')['downtime_hours'].sum().sort_values()
downtime_site
```

```python
plt.figure(figsize=(8,5))
downtime_site.plot(kind='barh')
plt.title('Horas de paro por área')
plt.xlabel('Horas de paro')
plt.ylabel('Área')
plt.tight_layout()
plt.show()
```

## Conclusión

El análisis permite identificar categorías con mayor volumen de órdenes, áreas con mayor costo, cumplimiento SLA y sitios con mayor tiempo de paro. Estos indicadores pueden utilizarse para priorizar mantenimiento preventivo y mejorar la gestión operativa.
